```python
moon = open('/content/drive/MyDrive/데이터분석입문/Doit_Python-main/Data/speech_moon.txt', encoding = 'UTF-8').read()
moon
```

```python
import re
moon = re.sub('[^가-힣]', ' ', moon)
moon
```

```python
import konlpy
hannanum = konlpy.tag.Hannanum()

hannanum.nouns("대한민국의 영토는 한반도와 그 부속도서로 한다")
```

```python
nouns = hannanum.nouns(moon)
nouns
```

```python
import pandas as pd
df_word = pd.DataFrame({'word' : nouns})
df_word
```

```python
df_word['count'] = df_word['word'].str.len()
df_word
```

```python
df_word = df_word.query('count >= 2')
df_word.sort_values('count')
```

```python
df_word = df_word.groupby('word', as_index = False).agg(n = ('word', 'count')).sort_values('n', ascending = False)
df_word
```

```python
top20 = df_word.head(20)
top20
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.rc('font', family='NanumGothic')
plt.rcParams.update({'figure.dpi' : '120',
                     'figure.figsize' : [6.5, 6]})

sns.barplot(data=top20, y='word', x='n')
```

#[10-2] 기사 댓글 텍스트 마이닝

```python
df = pd.read_csv('/content/drive/MyDrive/데이터분석입문/Doit_Python-main/Data/news_comment_BTS.csv', encoding='UTF-8')

df.info()
```

```python
df['relpy'] = df['reply'].str.replace('[^가-힣]', ' ', regex=True)
df['reply'].head()
```

```python
import konlpy
kkma = konlpy.tag.Kkma()
```

```python
nouns = df['reply'].apply(kkma.nouns)
nouns
```

```python
nouns = nouns.explode()
nouns
```

```python
df_word = pd.DataFrame({'word' : nouns})

df_word['count'] = df_word['word'].str.len()

df_word = df_word.query('count >= 2')
df_word
```

```python
df_word = df_word.groupby('word', as_index=False).agg(n=('word', 'count')).sort_values('n', ascending = False)
df_word
```

```python
top20 = df_word.head(20)
top20
```

```python
plt.rcParams.update({'figure.figsize' : [6.5,6]})
sns.barplot(data=top20, y = 'word', x= 'n')
```

```python
dic_word = df_word.set_index('word').to_dict()['n']
```

```python
wc = WordCloud(random_state=1234,
               font_path='/content/drive/MyDrive/데이터분석입문/Doit_Python-main/Data/DoHyeon-Regular.ttf',
               width=400,
               height=400,
               background_color='white',
               mask=img)
```

```python
img_wordcloud = wc.generate_from_frequencies(dic_word)

plt.figure(figsize=(10,10))
plt.axis('off')
plt.imshow(img_wordcloud)
```
